```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense, Dropout , LSTM
from keras import regularizers
```

```python
data = pd.read_csv('/content/ECG-Dataset.csv')
```

```python
data.columns = ['age','sex','smoker','years_of_smoking','LDL_cholesterol','chest_pain_type','height','weight', 'familyhist',
                'activity', 'lifestyle', 'cardiac intervention', 'heart_rate', 'diabets', 'blood_pressure_sys', 'blood_pressure_dias',
                 'hypertention', 'Interventricular_septal_end_diastole', 'ecg_pattern', 'Q_wave', 'target']
print(data)
print(data.shape)
```

```python
data.describe()
```

```python
# Total Percentage of Missing Data
missing_data = data.isnull().sum()
total_percentage = (missing_data.sum()/data.shape[0]) * 100
print(f'The total percentage of missing data is {round(total_percentage,2)}%')
```

```python
fig = plt.figure(figsize = (15,20))
ax = fig.gca()
data.hist(ax = ax)
plt.show()
```

```python
pd.crosstab(data.age,data.target).plot(kind="bar",figsize=(20,6))
plt.title('Heart Disease Frequency for Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

Data Preprocessing

```python
X = data.drop(columns=['target']).values
y = data['target'].values

# Normalize the data
mean = X.mean(axis=0)
X -= mean
std = X.std(axis=0)
X /= std
print(X)
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=35)
```

Artificial Neural Network Model

```python
model = Sequential()
model.add(Dense(64, input_dim=20, kernel_initializer='normal', kernel_regularizer=regularizers.l2(0.001), activation='relu'))
model.add(Dropout(0.20)) #to prevent overfitting
model.add(Dense(32, kernel_initializer='normal', kernel_regularizer=regularizers.l2(0.001), activation='relu'))
model.add(Dropout(0.20))
model.add(Dense(16, kernel_initializer='normal', kernel_regularizer=regularizers.l2(0.001), activation='relu'))
model.add(Dropout(0.20))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])

# train the modelm
history= model.fit(X_train, y_train, epochs=50, batch_size=32, validation_split=0.2)

# valuate the model
acc= model.evaluate(X_test, y_test)
print(f"Artificial Neural Network - Loss: {acc[0]*100:.4f}, Accuracy: {acc[1]*100:.4f}")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Model accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()
```

Recurrent Neural Network Model

```python

# RNN (LSTM)
X_train_rnn = X_train.reshape(X_train.shape[0], 1, X_train.shape[1])
X_test_rnn = X_test.reshape(X_test.shape[0], 1, X_test.shape[1])

model_rnn = Sequential()
model_rnn.add(LSTM(32, activation='relu', input_shape=(1, X_train.shape[1])))
model_rnn.add(Dropout(0.2))
model_rnn.add(Dense(1, activation='sigmoid'))

# Compile the model
model_rnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# train the model
history=model_rnn.fit(X_train_rnn, y_train, epochs=50, batch_size=32, validation_split=0.2)

# evaluate the model
rnn_loss, rnn_accuracy = model_rnn.evaluate(X_test_rnn, y_test)
print(f"Recurrent Neural Network - Loss: {rnn_loss*100:.4f}, Accuracy: {rnn_accuracy*100:.4f}")
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
# Model accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'])
plt.show()
```
